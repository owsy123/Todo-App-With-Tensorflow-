# Importing Libraries 

```python
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler 
import sys
np.set_printoptions(threshold=sys.maxsize)
```

# Data

```python
data = pd.read_excel("todoData.xlsx", date_parser=True)
data.head()
```

# Training Data

```python
#train data before 2019-06-30

training_data = data[data["Date"]< "2019-06-30"].copy()
training_data
```

# Testing Data 

```python
#test data after 2019-06-29

testing_data = data[data["Date"]>"2019-06-29"].copy()
testing_data
```

# Removing Date Columns From Train Data

```python
#removing unnecessary columns

train_data = training_data.drop(["Date"], axis=1)
train_data
```

# Normalizing Values

```python
#normalizing values 

scaler = MinMaxScaler()
train_data = scaler.fit_transform(train_data)
train_data
```

# Preparing Training Data 

```python
x_train = []
y_train = []
```

```python
# training on 3 days then predciting 4th value

for i in range (3, train_data.shape[0]):
    x_train.append(train_data[i-3:i])
    y_train.append(train_data[i])
    
# train_data[i]
```

```python
x_train , y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Building Model

```python
# building LSTM network

from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

regressor = Sequential()

regressor.add(LSTM(units = 60, activation = "relu", return_sequences = True, input_shape= (x_train.shape[1],29)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 60, activation = "relu", return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 80, activation = "relu", return_sequences=True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 120, activation = "relu"))
regressor.add(Dropout(0.2))
regressor.add(Dense(29, activation='sigmoid'))
```

```python
regressor.summary()
```

```python
#compiling the regressor model

regressor.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])
```

```python
#training the the model

regressor.fit(x_train, y_train, epochs=30, batch_size=12)
```

# Preparing Test Data 

```python
# preparing test data 

# we need to have last 3 days of training data to predict the first test data

past_3_days = training_data.tail(3)

df = past_3_days.append(testing_data, ignore_index = True)

# df
```

## Removing Date Columns From Test Data

```python
df = df.drop(["Date"],axis=1)
# df
```

# Normalizing Values

```python
inputs  = scaler.transform(df)
inputs
```

```python
x_test = []
y_test = []
```

```python
for i in range(3, inputs.shape[0]):
    x_test.append(inputs[i-3:i])
    y_test.append(inputs[i])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
x_test.shape, y_test.shape
```

# Prediction

```python
# predicting item
y_pred = regressor.predict(x_test)
```
